```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(-5,5,1000)
```

```python
mod_x=abs(x)
```

```python
plt.plot(x,mod_x,'b--',label="$|x|$",lw=2)
plt.legend()
plt.show()
```

```python
x_sq=x**2
x_cube=x**3
x_exp=np.exp(x)
x_log=np.log(x)
x_sqrt=np.sqrt(x)
```

```python
plt.plot(x,x_sq,label="Square")
plt.plot(x,x_cube,label="Cube")
plt.plot(x,x_exp,label="exponential")
plt.plot(x,x_log,label="logarithmic")
plt.plot(x,x_sqrt,label="square root")
plt.legend()
plt.show()
```

```python
y=np.sqrt(100-x_sq)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(5,10))
angle=np.arange(0,361,15)
rad=np.deg2rad(angle)
sin=np.sin(rad)
cos=np.cos(rad)
plt.plot(angle,sin,'g--',label="Sin",lw=3);plt.legend()
plt.plot(angle,cos,'r-',label="Cos",lw=3);plt.legend()
plt.title("Sin And Cos",fontsize=20)
plt.xlabel("Angle",fontsize="20")
plt.ylabel("Values",fontsize="20")

plt.grid()
plt.savefig("two.png")
plt.show()
```
