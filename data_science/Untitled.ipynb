```python
import numpy as np
import matplotlib.pyplot as plt
import psutil
from matplotlib.animation import FuncAnimation
```

```python

%matplotlib  notebook

fig=plt.figure()
ax=fig.add_subplot(111)
x=[var for var in range(50)]
y=[ psutil.cpu_percent() for var in range(50)]
ax.plot(x,y)
def change(i):
    global x,y
    x.pop(0)
    y.pop(0)
    x.append(x[-1]+1)
    y.append(psutil.cpu_percent())
    ax.clear()
    ax.plot(x,y,'r^--')
ani=FuncAnimation(fig,change,2000)
plt.show()
```
